```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
over=[5,10,15,20]
runs=[45,79,145,234]
runrate=np.divide(runs,over)
print(runrate)

```

```python
plt.title("Run Rate")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.plot(over,runs,color='cyan',marker='^',markersize=20,label="Runrate")

plt.legend()
plt.show()
```
