```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,3*np.pi,0.1)
x
```

```python
y=np.sin(x)
y
```

```python
plt.title("SIN graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='orange',markersize=10,label="graph")

plt.legend()
plt.show()
```

```python
y=np.cos(x)
plt.title("COSINE graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='cyan',markersize=10,label="graph")

plt.legend()
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,1)
plt.plot(x,z)
plt.show()
```
