```python
from matplotlib import pyplot as plt
```

```python
x1=[10,12,15]
y1=[12,22,23]
x2=[6,9,11]
y2=[6,16,18]
plt.bar(x1,y1,align='center')
plt.bar(x2,y2,align='center')
plt.show()
```

```python
x=[40,30,30]
label=['cyan','blue','red']
colors=['cyan','blue','red']
plt.pie(x,labels=label,explode=[1,0,0],colors=colors)

plt.show()
```
