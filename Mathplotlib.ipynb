```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(1,11)
x
```

```python
y=3*x+5
y
```

```python
plt.title("new graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='orange',marker='o',markersize=20,label="graph")

plt.legend()
plt.show()
```

```python
x=[1972,1980,1983,1993,2001]
y=[100,159,300,359,400]
z=[10,26,58,110,126]
```

```python
plt.title("Mixed")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='red',marker='o',markersize=10,label="xy")

plt.legend()


plt.xlabel("x axis")
plt.ylabel("z axis")
plt.plot(x,z,color='cyan',marker='*',markersize=10,label="xz")

plt.legend()
plt.show()
```
