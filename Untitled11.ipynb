```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,6,18]
y=[3,10,12,20]
plt.plot(x,y,'r--',marker="o",markerfacecolor="blue")
plt.grid(True)
```

```python
x1=[12,14,16,18,20,22,24]
y1=[100,200,250,400,300,450,500]
plt.plot(x1,y1,'r--',marker="o",markerfacecolor="blue")
plt.title('Temp vs Sales')
plt.xlabel('Sales')
plt.ylabel('Temperature in degree celsius')
plt.grid(True)
plt.show()
```
